```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

binary_array = np.load(r"D:\IceDyno\IMS_images\2015\npy\ims2015001_1km_v1_grid1000.npy")
```

```python
def find_edges(matrix):
    edges = []
    rows = len(matrix)
    cols = len(matrix[0])

    # Check horizontal edges
    for i in range(rows):
        for j in range(cols - 1):
            if matrix[i][j] != matrix[i][j + 1]:
                edges.append(((i, j), (i, j + 1)))

    # Check vertical edges
    for j in range(cols):
        for i in range(rows - 1):
            if matrix[i][j] != matrix[i + 1][j]:
                edges.append(((i, j), (i + 1, j)))

    return edges


edges = find_edges(binary_array)

plt.figure(figsize=(20, 20))
plt.imshow(binary_array, cmap="binary", interpolation="nearest")

# all_edges = []
for edge in edges:
    plt.plot(
        [edge[0][1], edge[1][1]], [edge[0][0], edge[1][0]], color="red", marker="."
    )


plt.title("Computed Edges")
plt.axis("off")
plt.show()
```
